To-Do 4: Cost Function (MSE)

```python
def cost_function(X, Y, W):
    """
    Mean Squared Error Cost Function
    """
    n = len(Y)
    Y_pred = np.dot(X, W)
    cost = (1 / (2 * n)) * np.sum((Y_pred - Y) ** 2)
    return cost
```

To-Do 6: Gradient Descent Implementation

```python
def gradient_descent(X, Y, W, alpha, iterations):
    """
    Gradient Descent for Linear Regression
    """
    cost_history = [0] * iterations
    m = len(Y)

    for i in range(iterations):
        # Step 1: Prediction
        Y_pred = np.dot(X, W)

        # Step 2: Loss
        loss = Y_pred - Y

        # Step 3: Gradient
        dw = (1 / m) * np.dot(X.T, loss)

        # Step 4: Update Weights
        W = W - alpha * dw

        # Step 5: Cost
        cost_history[i] = cost_function(X, Y, W)

    return W, cost_history
```

To-Do 7: Test Gradient Descent

```python
np.random.seed(0)

X_rand = np.random.rand(100, 3)
Y_rand = np.random.rand(100)
W_rand = np.random.rand(3)

alpha = 0.01
iterations = 1000

final_params, cost_history = gradient_descent(
    X_rand, Y_rand, W_rand, alpha, iterations
)

print("Final Parameters:", final_params)
print("Final Cost:", cost_history[-1])
```

To-Do 8: RMSE Implementation

```python
def rmse(Y, Y_pred):
    return np.sqrt(np.mean((Y - Y_pred) ** 2))
```

To-Do 9: R (Coefficient of Determination)

```python
def r2(Y, Y_pred):
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    return 1 - (ss_res / ss_tot)
```

To-Do 10: Main Function (Full Pipeline)

```python
def main():
    # Load data
    student = pd.read_csv("student.csv")

    # Features & target
    X = student[['Math', 'Reading']].values
    Y = student['Writing'].values

    # Train-test split
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=0.2, random_state=42
    )

    # Initialize parameters
    W = np.zeros(X_train.shape[1])
    alpha = 0.00001
    iterations = 1000

    # Train model
    W_optimal, cost_history = gradient_descent(
        X_train, Y_train, W, alpha, iterations
    )

    # Predictions
    Y_pred = np.dot(X_test, W_optimal)

    # Evaluation
    model_rmse = rmse(Y_test, Y_pred)
    model_r2 = r2(Y_test, Y_pred)

    # Output
    print("Final Weights:", W_optimal)
    print("Cost History (First 10):", cost_history[:10])
    print("RMSE:", model_rmse)
    print("R2 Score:", model_r2)
```

```python
if __name__ == "__main__":
    main()
```
